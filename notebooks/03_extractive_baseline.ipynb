Goal:

Pick the most important sentences from the article to form a summary.
Why extractive summarization is powerful

Instead of generating new text, we:


*   Select important sentences



*   Keep them as-is


*   Preserve factual correctness


This avoids Hallucinations, Grammar errors

That’s why extractive is used in:

News summarization, Legal documents, Medical reports

```python
import nltk

nltk.download("punkt")
nltk.download("stopwords")
nltk.download("wordnet")
nltk.download("omw-1.4")
nltk.download("punkt_tab")
```

```python
from datasets import load_dataset
from src.preprocess import preprocess_article

dataset = load_dataset("cnn_dailymail", "3.0.0")
article = dataset["train"][10]["article"]

original_sentences, cleaned_sentences = preprocess_article(article)
```

TF-IDF:

High score = rare but important words

Sentences containing many important words → important

```python
from src.extractive import tfidf_sentence_scores, get_top_sentences

scores = tfidf_sentence_scores(cleaned_sentences)
summary = get_top_sentences(original_sentences, scores, k=5)

print("ARTICLE:\n", article[:800])
print("\nSUMMARY:\n", summary)
```
