```python
import nltk

nltk.download("punkt")
nltk.download("stopwords")
nltk.download("wordnet")
nltk.download("omw-1.4")
nltk.download("punkt_tab")
```

```python
from datasets import load_dataset

# Load the CNN / DailyMail summarization dataset
dataset = load_dataset("cnn_dailymail", "3.0.0")

# Get one example (4th article from the training set)
sample = dataset["train"][3]

# Extract the news article text
article_text = sample["article"]

# Print only the first 1000 characters so it doesn't flood the screen
print(article_text[:1000])
```

```python
from nltk.tokenize import sent_tokenize

sentences = sent_tokenize(article_text)

print("Number of sentences:", len(sentences))
sentences[:5]
```

```python
from src.preprocess import preprocess_article

original, cleaned = preprocess_article(article_text)

for i in range(3):
    print("ORIGINAL:", original[i])
    print("CLEANED :", cleaned[i])
    print()
```

```python
import pandas as pd

df = pd.DataFrame({
    "original_sentence": original,
    "cleaned_sentence": cleaned
})

df.to_csv("data/processed/sample_sentences.csv", index=False)
```
